#  Predicciones con competition

```python
#Imports necesarios
import pandas as pd
import pickle
from joblib import load
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datos = pd.read_csv("attrition_competition_06.csv.gz", compression="gzip", sep=",")
```

```python
# Cargar el modelo
modelo_final = load("modelo_final_100496633_100496072.pkl")

# Eliminar la columna "Attrition" si existe
datos = datos.drop(columns=["Attrition"], errors='ignore')

# Convertir a DataFrame
datos = pd.DataFrame(datos)

# Hacer las predicciones
prediccion = modelo_final.predict(datos)
predicciones = pd.DataFrame(prediccion, columns=['Predicciones'])

# Guardar las predicciones en un archivo CSV
predicciones.to_csv('predicciones_100496633_100496072.csv', index=False)

# Imprimir la distribución de las predicciones
for i in predicciones.columns:
    print(f"Key: {i}\n{predicciones[i].value_counts(normalize=True) * 100}\n")
```

```python
# Calcular las proporciones
proporciones = predicciones['Predicciones'].value_counts(normalize=True) * 100

# Crear la figura con un fondo gris claro
fig, ax = plt.subplots(figsize=(14, 6), facecolor='whitesmoke')
ax.set_facecolor('lightgray')  # Fondo dentro de los ejes

# Contar las ocurrencias de cada clase
counts = predicciones['Predicciones'].value_counts()

# Colores personalizados para las barras
colors = ['#FF6F61', '#6B5B95']  # Rojo coral y azul violeta

# Visualización de las barras con colores personalizados
ax.bar(counts.index, counts.values, color=colors, edgecolor='black', linewidth=1.2)

# Etiquetas y títulos
ax.set_xlabel('Predicciones', fontsize=14, color='darkslategray')
ax.set_ylabel('Frecuencia', fontsize=14, color='darkslategray')
ax.set_title('Distribución de las Predicciones', fontsize=16, fontweight='bold', color='darkslategray')

# Etiquetar las barras con los porcentajes y valores
for p in ax.patches:
    height = p.get_height()
    width = p.get_width()
    x = p.get_x() + width / 2
    y = p.get_height()
    percentage = f'{y / len(predicciones) * 100:.2f}%'

    # Agregar un margen extra en el valor de y (y + 2) para el texto
    ax.text(x, y + 2, f'({height} = {percentage})', ha='center', fontsize=14, color='black', fontweight='bold')

# Establecer las etiquetas del eje X como 0 y 1
ax.set_xticks([0, 1])

# Mejorar la legibilidad de las etiquetas del eje X
plt.xticks(fontsize=12, rotation=0, ha='center', color='darkslategray')

# Mejorar la cuadrícula
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar el diseño para añadir margen
plt.tight_layout(pad=3.0)

# Ajustar el margen para centrar la gráfica
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

# Mostrar la gráfica
plt.show()

```
